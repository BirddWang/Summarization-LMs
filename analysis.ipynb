```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
scores = []
with open('data/cnndm_sumllm/gpt4/consistency_results.jsonl') as f:
    lines = f.readlines()
    for line in lines:
        data = json.loads(line)
        scores.append(data['score'])
```

```python
np.mean(scores)
```

```python
np.std(scores)
```

```python
plt.hist(scores)
```

```python
plt.boxplot(scores, showmeans=True, showbox=True, showcaps=True, showfliers=True)
```
